```python
from sklearn import datasets as dts
```

```python
from matplotlib import pyplot as plt
```

generating data

```python
x,y = dts.make_blobs(n_features = 2)
```

```python
#generated data
print(x)
print(y)
#slicing
print(x[:,0])
```

plotting data

```python
plt.scatter(x[:,0], x[:,1])
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
x,y = dts.make_blobs(n_features = 3)
```

```python
plt.scatter(x[:,0], x[:,1])
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

generating circle data

```python
x,y = dts.make_circles(n_samples = 100)
```

```python
plt.scatter(x[:,0],x[:,1],color = 'r')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
